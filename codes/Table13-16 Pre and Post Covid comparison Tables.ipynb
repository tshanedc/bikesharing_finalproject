```python
# import dependencies
import os
import pandas as pd

# read the all_bike_trips.csv file
filepath1 = '../tables/Table7_member_trips.csv'
member_df = pd.read_csv(filepath1, low_memory=False)
filepath2 = '../tables/Table8_casual_trips.csv'
casual_df = pd.read_csv(filepath2, low_memory=False)
```

```python
member_df = member_df[['startsstationnumber','startdate','membertype']]
member_df['WEEKDAY'] = ((pd.DatetimeIndex(member_df.startdate).dayofweek) // 5 == 1).astype(float)
member_df
```

```python
casual_df = casual_df[['startsstationnumber','startdate','membertype']]
casual_df['WEEKDAY'] = ((pd.DatetimeIndex(casual_df.startdate).dayofweek) // 5 == 1).astype(float)
casual_df
```

```python
# Create a dataframe of 2019-2020 Member
start_pre = '2019-01-01 00:01'
end_pre = '2019-12-31 23:59'
pre_covid = (member_df['startdate'] > start_pre) & (member_df['startdate'] < end_pre)
member_pre = member_df.loc[pre_covid].reset_index(drop=True)
member_pre
```

```python
# Create a dataframe of 2020-2021 Member
start_post = '2020-01-01 00:01'
end_post = '2021-12-31 23:59'
post_covid = (member_df['startdate'] > start_post) & (member_df['startdate'] < end_post)
member_post = member_df.loc[post_covid].reset_index(drop=True)
member_post
```

```python
# Create a dataframe of 2019-2020 Casual
start_pre = '2019-01-01 00:01'
end_pre = '2019-12-31 23:59'
pre_covid = (casual_df['startdate'] > start_pre) & (casual_df['startdate'] < end_pre)
casual_pre = casual_df.loc[pre_covid].reset_index(drop=True)
casual_pre
```

```python
# Create a dataframe of 2020-2021 Casual
start_post = '2020-01-01 00:01'
end_post = '2021-12-31 23:59'
post_covid = (casual_df['startdate'] > start_post) & (casual_df['startdate'] < end_post)
casual_post = casual_df.loc[post_covid].reset_index(drop=True)
casual_post
```

```python
# Convert the table into multi-index dataframe with startsstationnumber grouping weekday and count its frequency
member_pre_count = pd.DataFrame(member_pre.groupby('startsstationnumber')['WEEKDAY'].value_counts().fillna(0))
member_pre_count = member_pre_count.rename(columns = {'WEEKDAY':'count'}).reset_index()
member_pre_count['status'] = 'pre_covid'
member_pre_count.tail()
```

```python
# Convert the table into multi-index dataframe with startsstationnumber grouping weekday and count its frequency
member_post_count = pd.DataFrame(member_post.groupby('startsstationnumber')['WEEKDAY'].value_counts().fillna(0))
member_post_count = member_post_count.rename(columns = {'WEEKDAY':'count'}).reset_index()
member_post_count['status'] = 'post_covid'
member_post_count.tail()
```

```python
# Convert the table into multi-index dataframe with startsstationnumber grouping weekday and count its frequency
casual_pre_count = pd.DataFrame(casual_pre.groupby('startsstationnumber')['WEEKDAY'].value_counts().fillna(0))
casual_pre_count = casual_pre_count.rename(columns = {'WEEKDAY':'count'}).reset_index()
casual_pre_count['status'] = 'pre_covid'
casual_pre_count.tail()
```

```python
# Convert the table into multi-index dataframe with startsstationnumber grouping weekday and count its frequency
casual_post_count = pd.DataFrame(casual_post.groupby('startsstationnumber')['WEEKDAY'].value_counts().fillna(0))
casual_post_count = casual_post_count.rename(columns = {'WEEKDAY':'count'}).reset_index()
casual_post_count['status'] = 'post_covid'
casual_post_count.tail()
```

```python
# Merge the 'member_pre_count' and 'casual_pre_count' together on 'startsstationnumber' and 'weekday'
merge_pre = member_pre_count.merge(casual_pre_count, 
                                how='outer', 
                                left_on=['status',"startsstationnumber", "WEEKDAY"], 
                                right_on=['status',"startsstationnumber", "WEEKDAY"]).fillna(0)
merge_pre = merge_pre[['status','startsstationnumber','WEEKDAY','count_x','count_y']]
merge_pre
```

```python
# Merge the 'member_post_count' and 'casual_post_count' together on 'startsstationnumber' and 'weekday'
merge_post = member_post_count.merge(casual_post_count, 
                                how='outer', 
                                left_on=['status',"startsstationnumber", "WEEKDAY"], 
                                right_on=['status',"startsstationnumber", "WEEKDAY"]).fillna(0)
merge_post = merge_post[['status','startsstationnumber','WEEKDAY','count_x','count_y']]
merge_post
```

```python
# Merge the pre-Covid member trip and casual trip dataframes and add ratio column
pre_ratio = merge_pre.assign(ratio = merge_pre['count_x'] / (merge_pre['count_x'] + merge_pre['count_y'])).round(2)
pre_ratio = pre_ratio.assign(trip_total = merge_pre['count_x'] + merge_pre['count_y'])
pre_ratio = pre_ratio[['status','startsstationnumber','WEEKDAY','ratio', 'trip_total',
                       'count_x','count_y']].rename(columns = {'count_x':'trip_by_member',
                                                               'count_y':'trip_by_casual'
                                                                })
pre_ratio
```

```python
# Merge the pre-Covid member trip and casual trip dataframes and add ratio column
post_ratio = merge_post.assign(ratio = merge_post['count_x'] / (merge_post['count_x'] + merge_post['count_y'])).round(2)
post_ratio = post_ratio.assign(trip_total = merge_post['count_x'] + merge_post['count_y'])
post_ratio = post_ratio[['status','startsstationnumber','WEEKDAY','ratio', 'trip_total',
                       'count_x','count_y']].rename(columns = {'count_x':'trip_by_member',
                                                               'count_y':'trip_by_casual'
                                                                })
post_ratio
```

```python
# Output to csv
folder_path = '../tables/'
pre_ratio.to_csv(os.path.join(folder_path,'Table13_pre_ratio.csv'),index=False)
folder_path = '../tables/'
post_ratio.to_csv(os.path.join(folder_path,'Table14_post_ratio.csv'),index=False)
```

```python
# merge pre and post covid dataframes
merge_covid = pd.concat([pre_ratio,post_ratio]).fillna(0)
merge_covid
```

```python
# Output to csv
folder_path = '../tables/'
merge_covid.to_csv(os.path.join(folder_path,'Table15_merge_covid.csv'),index=False)
```

```python
# Figure out the total member vs. casual ratio of pre-Covid
pre_count = pd.DataFrame({'Pre_Covid': pre_ratio[['trip_total','trip_by_member','trip_by_casual']].sum(axis=0)}) 
pre_count = pre_count.T
pre_count = pre_count.assign(ratio = pre_count['trip_by_member'] / pre_count['trip_total'])
pre_count 
```

```python
# Figure out the total member vs. casual ratio of post-Covid
post_count = pd.DataFrame({'Post_Covid': post_ratio[['trip_total','trip_by_member','trip_by_casual']].sum(axis=0)}).T 
post_count = post_count.assign(ratio = post_count['trip_by_member'] / post_count['trip_total'])
post_count 
```

```python
pre_post_total = pd.concat([pre_count,post_count], axis=0).reset_index()
pre_post_total.head()
```

```python
# Output to csv
folder_path = '../tables/'
pre_post_total.to_csv(os.path.join(folder_path,'Table16_pre_post_total.csv'),index=False)
```
