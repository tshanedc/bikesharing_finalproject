```python
# import dependencies
import os
import pandas as pd
```

```python
# read the all_bike_trips.csv file
filepath = '../Datasets/Washington DC/tables/Table1_all_bike_trips.csv'
df = pd.read_csv(filepath, low_memory=False)
df.head()
```

```python
# separate the data by Member users and Casual users
df_member = df[(df['membertype'] == 'Member')]
df_casual = df[(df['membertype'] == 'Casual')]
df_member
```

```python
df_casual
```

```python
# Output to csv
folder_path = '../Datasets/Washington DC/tables/'
df_member.to_csv(os.path.join(folder_path,'Table7_member_trips.csv'),index=False)
folder_path = '../Datasets/Washington DC/tables/'
df_casual.to_csv(os.path.join(folder_path,'Table8_casual_trips.csv'),index=False)
```

```python
# Create a dataframe of stationnumber with trips taken by members
member_count = pd.DataFrame(df_member['startsstationnumber'].value_counts())
member_count = member_count.rename(columns={'startsstationnumber':'number_of_members'})
member_count['startsstationnumber'] = member_count.index
member_count = member_count[['startsstationnumber','number_of_members']]
member_count = member_count.reset_index(drop=True)
member_count.dtypes
```

```python
# Create a dataframe of stationnumber with trips taken by casuals
casual_count = pd.DataFrame(df_casual['startsstationnumber'].value_counts())
casual_count = casual_count.rename(columns={'startsstationnumber':'number_of_casuals'})
casual_count['startsstationnumber'] = casual_count.index
casual_count = casual_count[['startsstationnumber','number_of_casuals']]
casual_count = casual_count.reset_index(drop=True)
casual_count.dtypes
```

```python
# Merge the member trip and casual trip dataframes and add ratio column
member_casual = pd.merge(member_count,casual_count, on = 'startsstationnumber', how = 'inner')
ratio_df = member_casual.assign(ratio = member_casual['number_of_members'] / (member_casual['number_of_members'] + member_casual['number_of_casuals'])).round(2)
ratio_df = ratio_df[['startsstationnumber','ratio','number_of_members','number_of_casuals']]
ratio_df = ratio_df.sort_values(by='startsstationnumber', ascending=True, na_position='first').reset_index(drop=True)
ratio_df
```

```python
# Output to csv
folder_path = '../Datasets/Washington DC/tables/'
ratio_df.to_csv(os.path.join(folder_path,'Table9_ratio_df.csv'),index=False)
```
