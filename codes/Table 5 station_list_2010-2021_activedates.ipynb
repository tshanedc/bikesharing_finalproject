```python
# import dependencies
import os
import pandas as pd
```

```python
# import index csv for datasets
#file_path = "../Datasets/Washington DC/capitalbikeshare_dataset_index.xlsx"
folder_path = '../Datasets/Washington DC/'
file_name = 'capitalbikeshare_dataset_index.csv'
index_df=pd.read_csv(f"{folder_path}{file_name}")
```

```python
# Adjust the "Year" column as string data to add 
index_df['Year'] = index_df['Year'].astype(str)
index_df.dtypes
```

```python
# Create another list of files without lat/lng info
# Create the list of csv file paths we want to read

csv_file_list = []
# loop through the range of index we are interested to create a list of file path of csv files
for i in range(0,52):
    Year = index_df['Year'][i]+'/'
    file_name = index_df['File Name'][i]+'.csv'
    file_name = f'{file_name}'
    folder_path = '../Datasets/Washington DC/'
    folder_path = f"{folder_path}{Year}{file_name}"
    csv_file_list.append(f"{folder_path}")
csv_file_list
```

```python
# Read the listed file paths as DataFrame and merge

list_of_datasets = []
# loop through pd.read_csv method with the list of file path we created earlier 
for filepath in csv_file_list:
    list_of_datasets.append(pd.read_csv(filepath, low_memory=False))

# merge the dataframes we read into one dataframe
merge_df = pd.concat(list_of_datasets)
merge_df
```

```python
# Drop unnecessary columns for geo data table and check data types
merge_df_dropped = merge_df[['Start station number','Start station','End station number','End station','Start date','End date']]
merge_df_dropped
```

```python
merge_df_dropped.dtypes
```

```python
# Identify any NaN in the dataset
merge_df_dropped.isnull().sum()
```

```python
# Convert the Start date and End date columns to datetime data type
merge_df_dropped['Start date'] = pd.to_datetime(merge_df_dropped['Start date'])
merge_df_dropped['End date'] = pd.to_datetime(merge_df_dropped['End date'])
merge_df_dropped.dtypes
```

```python
# Split data
start_stations = merge_df_dropped[['Start station number','Start station','Start date','End date']]
end_stations = merge_df_dropped[['End station number','End station','Start date','End date']]
```

```python
# Rename columns to remove 'start' and 'end' then merge them together
start_stations = start_stations.rename(columns={'Start station number':'station_id','Start station':'station_name'})
end_stations = end_stations.rename(columns={'End station number':'station_id','End station':'station_name'})
frames = [start_stations,end_stations]
station_list = pd.concat(frames)
station_list_earlier = station_list.sort_values(by='station_id', ascending=True, na_position='first')
station_list_earlier 
```

```python
# Create the list of csv file paths we want to read
csv_file_list = []
# loop through the range of index we are interested to create a list of file path of csv files
for i in range(53,65):
    Year = index_df['Year'][i]+'/'
    file_name = index_df['File Name'][i]+'.csv'
    file_name = f'{file_name}'
    folder_path = '../Datasets/Washington DC/'
    folder_path = f"{folder_path}{Year}{file_name}"
    csv_file_list.append(f"{folder_path}")
csv_file_list
```

```python
# Read the listed file paths as DataFrame and merge
list_of_datasets = []
# loop through pd.read_csv method with the list of file path we created earlier 
for filepath in csv_file_list:
    list_of_datasets.append(pd.read_csv(filepath, low_memory=False))

# merge the dataframes we read into one dataframe
merge_df = pd.concat(list_of_datasets)
merge_df
```

```python
# Drop unnecessary columns for geo data table and check data types
merge_df_dropped = merge_df[['start_station_id','end_station_id','started_at','ended_at']]
merge_df_dropped
```

```python
# Identify any NaN in the dataset
merge_df_dropped.isnull().sum()
```

```python
# Take rows of 'start_station_id' and 'end_station_id' which are not NaN.
merge_cleaned = merge_df_dropped[merge_df_dropped['start_station_id'].notna()]
merge_cleaned = merge_cleaned[merge_cleaned['end_station_id'].notna()]
merge_cleaned.isnull().sum()
```

```python
# Remove a row with mixed data 
merge_cleaned = merge_cleaned[~(merge_cleaned == "MTL-ECO5-03").any(axis=1)]
```

```python
# Convert the 'started_at' and 'ended_at' columns to datetime data type
merge_cleaned['started_at'] = pd.to_datetime(merge_cleaned['started_at'])
merge_cleaned['ended_at'] = pd.to_datetime(merge_cleaned['ended_at'])
merge_cleaned['start_station_id'] = merge_cleaned['start_station_id'].astype(int)
merge_cleaned['end_station_id'] = merge_cleaned['end_station_id'].astype(int)
merge_cleaned.dtypes
```

```python
# Split data
start_stations = merge_cleaned[['start_station_id','started_at','ended_at']]
end_stations = merge_cleaned[['end_station_id','started_at','ended_at']]
# Rename columns to remove 'start' and 'end' then merge them together
start_stations = start_stations.rename(columns={'start_station_id':'station_id'})
end_stations = end_stations.rename(columns={'end_station_id':'station_id'})
frames = [start_stations,end_stations]
station_list = pd.concat(frames)
station_list_later = station_list.sort_values(by='station_id', ascending=True, na_position='first')
station_list_later
```

```python
# match the format of station_list_earlier with station_list_later
station_list_earlier = station_list_earlier[['station_id','Start date','End date']]
station_list_earlier = station_list_earlier.rename(columns={'Start station number':'station_id','Start date':'started_at','End date':'ended_at'})
station_list_earlier
```

```python
# merge the 2010-2020/03 and 2020/5-2021/05 bike trips together
frames = [station_list_earlier ,station_list_later]
station_list_all = pd.concat(frames)
station_list_all = station_list_all.sort_values(by='started_at', ascending=True, na_position='first')
station_list_all
```

```python
# From the 56M entries of trips, pick the first day and the last day
station_dates = station_list_all.groupby(['station_id'])['started_at'].agg(['first','last'])
station_dates
```

```python
station_list = station_list_all['station_id'].drop_duplicates(keep='last')
station_id = station_list.sort_values(ascending=True).reset_index(drop=True)
station_id
```

```python
# add 'station_id' as first column
df = station_dates.assign(startstationnumber=station_id.values)
station_active_dates = df[['station_id','first','last']].reset_index(drop=True)
station_active_dates
```

```python
# Output to csv
folder_path = '../Datasets/Washington DC/tables/'
station_active_dates.to_csv(os.path.join(folder_path,'Table5_station_active_dates.csv'),index=False)
```
