```python
# import dependencies
import os
import pandas as pd
```

```python
# import index csv for datasets
#file_path = "../Datasets/Washington DC/capitalbikeshare_dataset_index.xlsx"
folder_path = '../Datasets/Washington DC/'
file_name = 'capitalbikeshare_dataset_index.csv'
index_df=pd.read_csv(f"{folder_path}{file_name}")
```

```python
# Adjust the "Year" column as string data to add 
index_df['Year'] = index_df['Year'].astype(str)
index_df.dtypes
```

```python
def read_files(i):
    # for i in range(0,1):
        Year = index_df['Year'][i]+'/'
        file_name = index_df['File Name'][i]+'.csv'
        file_name = f'{file_name}'
        folder_path = '../Datasets/Washington DC/'
        folder_path = f"{folder_path}{Year}{file_name}"
        df = pd.read_csv(f"{folder_path}")
        return df
df = read_files(62)
df
```

```python
# Drop unnecessary columns for geo data table and check data types
df = df.drop(columns = ['ride_id','rideable_type','started_at','ended_at','member_casual'])
df.dtypes
```

```python
# Identify any NaN in start_station_id
df['start_station_id'].isnull().sum()
```

```python
# Take rows of 'start_station_id' and 'start_lat' which are not NaN.
df = df[df['start_station_id'].notna()]
df = df[df['start_lat'].notna()]
df['start_station_id'].isnull().sum()
```

```python
# convert to integer
df['start_station_id'] = pd.to_numeric(df['start_station_id'])
df['end_station_id'] = pd.to_numeric(df['end_station_id'])
df.dtypes
```

```python
# Organize the table by columns and sort by 'start_station_id'
df = df[['start_station_id','end_station_id','start_station_name','end_station_name','start_lat','start_lng','end_lat','end_lng']]
df = df.sort_values(by='start_station_id', ascending=True, na_position='first')
df
```
